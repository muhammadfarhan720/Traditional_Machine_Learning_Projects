```python
def tryVariableSelection(k):
    import pandas as pd

    import sklearn.cluster as cluster



    clus = cluster.KMeans(n_clusters=k,random_state=42)

    ans = clus.fit_predict(iris_x)


    iris_dataFrame['labels'] = ans

    #dataFrame.to_excel( fileName.replace('.xlsx', 'labeled.xlsx'))



    iris_dataFrame.to_excel("New_iris.xlsx", index=True)

    #**print(iris_x)

    #X = df.to_numpy()
    # Apply K-Mean Cluster
    #clus = cluster.KMeans(n_clusters=2)
    #labels = clus.fit_predict(X)
    #df['cluster'] = labels

    from scipy.stats import norm

    centroids = clus.cluster_centers_

    #***print(centroids)
    y= centroids[iris_dataFrame['labels']]

    #**print("y")

    #**print(y)

    #**print("iris_x")

    #**print(iris_x)

    #initializing two arrays
    array1 = iris_x
    array2 = y

    #computing the Euclidan distance
    temp = array1 - array2
    dista = np.linalg.norm(temp, axis=1)
    a=dista


    #**print("distance of each point and it's centroid")
    #**print(a)
    #**print(a.size)




    from scipy.spatial import distance

    Y = distance.cdist(y,centroids , 'euclidean')

    #**print("Distance of one points centroid to other centroids")


    Y[np.where(Y==0)]=100
    #***print(Y)

    min_element = np.min(Y,1)

    b=min_element

    #**print(b)
    sz=min_element.size
    #***print(sz)
    sum=0
    for i in range(sz):

      if a[i]<b[i]:
        values=1-a[i]/b[i]

      elif a[i]==b[i]:
        values=0
      elif a[i]>b[i]:
        values=b[i]/a[i]-1

      sum=sum+values


    silhouette_score=sum/sz
    print("condensed_silhouette for iris_data k=%f"%k)
    print(silhouette_score)

for k in range(2,21,1):
   tryVariableSelection(k)



#distance.cdist(iris_x[])
#drop_iris=iris_x.drop(["labels"],axis=1)
#print(iris_x)
#dist=  np.linalg.norm(iris_x[0,:]-y[0])
#print(dist)

#import math
#edistance=math.dist(iris_x,y)
#print(edistance)
#dista = sdist.norm(iris_x - y, axis=1)
#iris_dataFrame['dista'] = dista
```

```python
def show_plot(X, y):    # from scikitlearn.org/stable/auto_examples/decomposition/plot_pca_vs_lda.html
    plt.figure()
    colors = ["navy", "turquoise"]
    lw = 2
    indices = [1, 0]
    target_names = [1, 0]
    for color, i, target_name in zip(colors, indices, target_names):
        plt.scatter(X[y == i, 1], X[y == i, 0], color=color, alpha=0.8, lw=lw, label=target_name)
    plt.legend(loc="best", shadow=False, scatterpoints=1)
    plt.title("Scatterplot of iris_dataset clusters (Normalized)")

show_plot(iris_x,ans)
```
